Null value analysis findings:
The columns that contains null value are:
- Offer
- Avg Monthly Long Distance Charges
- Multiple Lines
- Internet Type
- Avg Monthly GB Download
- Online Security
- Online Backup
- Device Protection Plan
- Premium Tech Support
- Streaming TV
- Streaming Movies
- Streaming Music
- Unlimited Data
- Churn Category
- Churn Reason

The total rows entries are 7043.

- Now for 'Offer' column, there are 6 acceptable values: None, Offer A, Offer B, Offer C, Offer D and Offer E as per dictionary table. Now in main table, there are these 5 offer values along with nulls. So, we need to replace these nulls with 'None'

- It is given in dictionary table that 'Avg Monthly Long Distance Charges' will be 0 and 'Multiple Lines' will be 'No' if customer is not subscribed to any home phone service, i.e., 'Phone Service' value is 'No'.

So, for 'Phone service' has total 682 'No' value counts.
And after checking data, we found that 'Avg Monthly Long Distance Charges' and 'Multiple Lines' have nulls exactly where 'Phone service' value is 'No'. 
So these total nulls needs to be replaced with either 0 or 'No' as per the column.

- It is given in dictionary table that 'Internet Type' will be 'None', 'Avg Monthly GB Download' will be 0 and 'Online Security', 'Online Backup', 'Device Protection Plan', 'Premium Tech Support', 'Streaming TV', 'Streaming Movies', 'Streaming Music' and 'Unlimited Data' will be 'No'if customer is not subscribed to internet service, i.e., 'Internet Service' value is 'No'.

And after checking data, we found that these columns have nulls exactly where 'Internet service' value is 'No'. 
So these total nulls needs to be replaced with 0, 'No' or 'None' as per the column.

- 'Churn Category' and 'Churn Reason' has nulls where 'Customer Status' is 'Stayed' or 'Joined' which is valid.

```python
customer_churn_df['Offer'] = customer_churn_df['Offer'].fillna('None')
customer_churn_df['Avg Monthly Long Distance Charges'] = customer_churn_df['Avg Monthly Long Distance Charges'].fillna(0)
customer_churn_df['Multiple Lines'] = customer_churn_df['Multiple Lines'].fillna('No')
customer_churn_df['Internet Type'] = customer_churn_df['Internet Type'].fillna('None')
customer_churn_df['Avg Monthly GB Download'] = customer_churn_df['Avg Monthly GB Download'].fillna(0)
customer_churn_df['Online Security'] = customer_churn_df['Online Security'].fillna('No')
customer_churn_df['Online Backup'] = customer_churn_df['Online Backup'].fillna('No')
customer_churn_df['Device Protection Plan'] = customer_churn_df['Device Protection Plan'].fillna('No')
customer_churn_df['Premium Tech Support'] = customer_churn_df['Premium Tech Support'].fillna('No')
customer_churn_df['Streaming Movies'] = customer_churn_df['Streaming Movies'].fillna('No')
customer_churn_df['Streaming Music'] = customer_churn_df['Streaming Music'].fillna('No')
customer_churn_df['Streaming TV'] = customer_churn_df['Streaming TV'].fillna('No')
customer_churn_df['Unlimited Data'] = customer_churn_df['Unlimited Data'].fillna('No')
```
